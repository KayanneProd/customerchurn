## Data Cleaning

```python
# Make a copy of the data for manipulation and processing
telco_data = raw_data.copy()
```

### TotalCharges

```python
# Total charges should be numeric. (In previous attempts I dropped this column, however in this attempt I will keep it)
telco_data.TotalCharges = pd.to_numeric(telco_data.TotalCharges, errors='coerce')
telco_data.isnull().sum()
```

There are 11 missing values for TotalCharges

```python
# Check records of missing values
telco_data.loc[telco_data['TotalCharges'].isnull() == True]
```

Since the amount of missing values is low (11 out of 7043), it is safe to just remove these records.

```python
# Remove Missing values
telco_data.dropna(how = 'any', inplace = True)
```

```python
# Check to see if records with null values have been removed
telco_data.isna().sum()
```

### Tenure

In previous attempts I left the tenure column as it was, however I will divide customers based on tenure for this attempt. I believe this will be better.

```python
# Max tenure
print(telco_data['tenure'].max())
```

72 months is 6 years. I can group customers into 6 different bins based on year. (Bins of 12 months)

```python
# Create labels for the bins
labels = ['{} - {}'.format(i, i + 11) for i in range(1,72,12)]

# Cut the data into bins
telco_data['tenure_group'] = pd.cut(telco_data.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
# See the value counts
telco_data['tenure_group'].value_counts()
```

```python
# Remove Unecessary columns
telco_data = telco_data.drop(['customerID', 'tenure'], axis=1)
```

```python
telco_data.head()
```

## Data Exploration

### Univariate Analysis

```python
for i, predictor in enumerate(telco_data.drop(['Churn', 'TotalCharges', 'MonthlyCharges'], axis=1)):
    plt.figure(i)
    sns.countplot(data=telco_data, x=predictor, hue='Churn')
```

### Convert Churn to binary

```python
# Map churn 'yes' and 'no' to 1 and 0 respectively. 
telco_data['Churn'] = telco_data['Churn'].map({'No':0, 'Yes':1})
```

```python
# Check the data
telco_data.head()
```

In previous attempts I did it manually, but now I have found out how to create all dummy variables automatically

```python
# automatically creates the dummy variables
telco_data_dummies = pd.get_dummies(telco_data)
telco_data_dummies.head()
```

### Relationship between Monthly Charges and Total Charges

```python
sns.lmplot(data=telco_data_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

Total Charges increase as Monthly Charges increase - expected

### Churn by Monthly Charges and Total Charges

```python
Mth = sns.kdeplot(telco_data_dummies.MonthlyCharges[(telco_data_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Mth = sns.kdeplot(telco_data_dummies.MonthlyCharges[(telco_data_dummies["Churn"] == 1) ],
                ax =Mth, color="Blue", shade= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly charges by churn')
```

Churn is high when Monthly Charges are high

```python
Tot = sns.kdeplot(telco_data_dummies.TotalCharges[(telco_data_dummies['Churn'] == 0)],
                  color='red', shade = True)
Tot = sns.kdeplot(telco_data_dummies.TotalCharges[(telco_data_dummies['Churn'] == 1)],
                  ax=Tot, color='blue', shade=True)
Tot.legend(['No churn', 'Churn'], loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

Higher churn at lower Total Charges

When taking into account: Tenure, Montly Charges, and Total Charges we see that Higher Monthly Charge at lower tenure results into lower Total Charge. Hence, all these 3 factors: Higher Monthly Charge, Lower tenure, and Lower Total Charge are linked to High Churn.

This insight is already much better than the insights my previous attempts have given me

### Bivariate Analysis

```python
# Put the records of churn into two variables, one for churned customers and another for non-churned customers
new_df1_target0 = telco_data.loc[telco_data['Churn'] == 0]
new_df1_target1 = telco_data.loc[telco_data['Churn'] == 1]
```

```python
# Instead of plotting all the tables manually, it is better to write a function to make it easier
def uniplot(df,col,title,hue = None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

```python
uniplot(new_df1_target1, col='Partner',title='Distribution of Gender for Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target0,col='Partner',title='Distribution of Gender for Non Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target1,col='PaymentMethod',title='Distribution of PaymentMethod for Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target1,col='Contract',title='Distribution of Contract for Churned Customers',hue='gender')
```
